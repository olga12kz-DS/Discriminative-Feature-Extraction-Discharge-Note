# **TEXT EXTRACTION AND CLEAN-UP**

From the full record, extract "Chief Complaint" and add it as a new column.

```python
# User Defined Function to extract Chief Complaint: (everything after "Chief Complaint:" until the end of the line) and add it as a column to the original dataframe.

def extract_complaint(pattern, df_discharge_text, column_name="text"):
    """
    Extracts matches based on a pre-defined regex pattern from a specified column in the dataframe.

    Parameters:
        pattern (str): The regular expression pattern to use for extraction: r'Chief Complaint:\s*(.*)'
        df (pd.DataFrame): The DataFrame containing the text column: df_discharge_text.
        column_name (str): The name of the column to search, "text_access".

    Returns:
        pd.Series: A Series containing the extracted matches which will be added as a new column to the original dataframe.
    """
    # applying the pre-defined regex to each row of the target column
    return df_discharge_text[column_name].apply(lambda text: re.search(pattern, str(text)).group(1) if re.search(pattern, str(text)) else None)

# df[column_name] → Selects the target column ("text_access" in df_discharge_text).
# .apply(lambda text: ...) → Applies a function to each row in that column.
# str(text) → Converts the value to a string (in case it contains NaN or numbers).
# re.search(pattern, str(text)) → Searches for the regex pattern inside each row’s text:
# If no match is found, re.search() returns None.
# If a match is found, .group(1) extracts the first capturing group from the regex.
# if re.search(pattern, str(text)) else None → Ensures that if no match is found, it returns None instead of causing an error.

# Define regex pattern
pattern = r'Chief Complaint:\s*(.*)'

# Apply function to extract chief complaints
df_extract["Chief Complaint"] = extract_complaint(pattern, df_discharge_text, "text")

# Display results
df_extract
```

From the full record, extract "Diagnosis" and add it as a new column.

```python
# User Defined Function to extract Diagnoses: (multi-line after "Discharge Diagnoses" until break or Discharge Condition:) and add it as a column to the original dataframe.

def extract_all_dz(pattern, df_discharge_text, column_name="text"):
    """
    Extracts matches based on a pre-defined regex pattern from a specified column in the dataframe.

    Parameters:
        pattern (str): The regular expression pattern to use for extraction:
        df (pd.DataFrame): The DataFrame containing the text column: df_discharge_text.
        column_name (str): The name of the column to search, "text_access".

    Returns:
        pd.Series: A Series containing the extracted matches which will be added as a new column to the original dataframe.
    """
    # applying the pre-defined regex to each row of the target column
    return df_discharge_text[column_name].apply(lambda text: re.search(pattern, str(text)).group(1) if re.search(pattern, str(text)) else None)


# Define regex pattern
pattern = pattern = r'(?s)Discharge Diagnosis:\s*(.+?)(Discharge Condition:|\Z)'

# \n\s*\n stops at the first blank line (double newline) to prevent capturing unwanted sections.
# \Z ensures it captures everything until the end if no blank line is found.

# Apply function to extract diagnosis
df_extract["All Dz"] = extract_all_dz(pattern, df_discharge_text, "text")

# Display results
df_extract
```

# **DISCRIMINATIVE FEATURE SELECTION**

Discriminative Feature Selection is a process of identifying the words, phrases, or features that best differentiate between two or more classes. In our case, the positive class is notes with chest pain complaints and the negative class is notes without such complaints. We will focus on the "Present History Illness" part of the data set to extract features differentiating the two classes.

The first method we will test is the Chi Square model which measures the strength of the relationship between a word (or n-gram) and the class label. Using class labels (1=positive class, 0 = negative class, in a binary prediction), it scores and ranks each feature (word/n-gram) on how strongly it is associated with either label.

Step 1: Review all chest-related complaints (to familiarize yourself with options).
Step 2: Extract all notes with the word "chest" in "Chief Complaint".
Step 3: Create a binary class column, Chest Complaint = 1, No Chest Complaint = 0 (qualification = presence of feature "chest").
Step 4: Run the feature discriminative selection using the Chi-square test.

```python
# first, we need to identify all chest-related complaints to familiarize ourselves with choices.
df_chest = df_extract[df_extract["Chief Complaint"].fillna("").str.contains(r"\bchest\b", case=False, regex=True)]
df_chest["Chief Complaint"].unique() # cannot see the full array

complaints = df_chest["Chief Complaint"].unique().tolist()
for c in complaints:
    print(c)
```

```python
# check records with no chief complaint
df_blank = df_extract[df_extract["Chief Complaint"] == ""]
```

```python
# Check for blanks and na in the target column.
(df_extract["Chief Complaint"] == " ").sum()
(df_extract["Chief Complaint"].isna() | (df_extract["Chief Complaint"].str.strip() == "")).sum()

# We can drop these blank records as they constitute less than 5% of the total.
df_extract.dropna(subset=["Chief Complaint"], inplace=True)
df_extract = df_extract[df_extract["Chief Complaint"].str.strip() != ""]
```

```python
# Doube-check blanks and na records.
(df_extract["Chief Complaint"].isna() | (df_extract["Chief Complaint"].str.strip() == "")).sum()
```

```python
# Now we will add the label column coded as a Boolean column, based on the presence of the word "chest".

# Define regex pattern for chest-related complaints. We will use only the word "chest" which appears to be the key word paired with secondary words like pain and pressure.
pattern = r"chest"
# Option 1: any non-string records will be assigned False or 0 (No Chest Pain)
# Apply case-insensitive search using the pre-defined regex pattern to the target column.
df_extract["Chest Pain Label"] = df_extract["Chief Complaint"] \
    .fillna("").str.lower().str.contains(pattern, regex=True).astype(int)

# Option 2:
# Create a full binary column with a lambda function to explicitly assign 0 when the pattern is not found
# (including blanks or NaN). Lambda functions are a great choice here as they are quick, local methods to iterate
# over elements with functions like apply() and map().

df_extract["Chest Pain Label"] = df_extract["Chief Complaint"].apply(
    lambda x: 1 if isinstance(x, str) and re.search(pattern, x, flags=re.IGNORECASE) else 0
)

df_chest_chi = df_extract[df_extract["Chest Pain Label"] ==  1]
df_chest_chi.head()
```

```python
# Use the Present Illness column as input text
target_text = df_extract["Present Illness History"].fillna("").tolist()

labels = df_extract["Chest Pain Label"]

# Vectorize using unigrams and bigrams, remove English stopwords
vectorizer = CountVectorizer(ngram_range=(1, 3), stop_words='english')
X = vectorizer.fit_transform(target_text)

# Compute chi-square scores
chi2_scores, _ = chi2(X, labels)
feature_names = vectorizer.get_feature_names_out()

# Get top 20 discriminative features
top_features = sorted(zip(chi2_scores, feature_names), reverse=True)[:20]

# Display in table
pd.DataFrame(top_features, columns=["Score", "Feature"])
```

The Chi-square score helps with separating high-signal features from high-noise features. The higher the score, the greater is the discriminating power of a given word or, in this case, a phrase of two words. For example, the feature "cath" with the score of 9,917 is strongly associated with the positive class, "Chest Complaint", i.e., it appears with a greater frequency in discharge notes with some type of chest complaint. The score is the measure of the difference between expected and observed frequences.

```python
# Compare to a method of limiting extracted and ranked features to only bi-grams, i.e., two-word phrases.
# This method gives additional insights because the word "chest" is usually paired with another word like "pressure", "pain", "discomfort".


target_text = df_extract["Present Illness History"].fillna("").tolist()
labels = df_extract["Chest Pain Label"]

# Vectorize using unigrams, bigrams, and trigrams
vectorizer = CountVectorizer(ngram_range=(1, 2), stop_words='english')
X = vectorizer.fit_transform(target_text)
feature_names = vectorizer.get_feature_names_out()

# Compute chi-square scores
chi2_scores, _ = chi2(X, labels)

# Combine scores and feature names
scored_features = list(zip(chi2_scores, feature_names))

# Filter for only bigrams (i.e., features with exactly one space)
bigram_features = [(score, feat) for score, feat in scored_features if feat.count(" ") == 1]

# Get top 20 bigrams by score
top_bigrams = sorted(bigram_features, reverse=True)[:100]

# Display
pd.DataFrame(top_bigrams, columns=["Score", "Feature"])
```
